```python
from common import get_prediction, split_data, get_data
import preprocessing as prep
import numpy as np
```

## Cargamos el dataset

```python
data = get_data()
data.head()
```

```python
pred = get_prediction()
pred.head()
```

## Dividimos los datos

Separamos un validation para los modelos

```python
X_train_val, X_val_2, y_train_val, y_val_2 = split_data(data, pred)
X_train_val.head()
```

Separamos otro validation para elegir el mejor modelo en una familia de modelos con el mejor preprocesado

```python
X_train, X_val, y_train, y_val = split_data(X_train_val, y_train_val)
X_train.head()
```

## Preprocesamos el dataset

El igual que los anteriores modelos se probara todos los preprocesados

```python
# Con la columna fila
prep_base_1 = prep.Preprocessing(
    [prep.Drop_high_cardinals(), prep.Dummy_variables()]
)

# Rellenado de nan con zeros
prep_1 = prep.Preprocessing(
    [prep_base_1, prep.Nan_to_zero()]
)

# Rellenado de nan con la media
prep_2 = prep.Preprocessing(
    [prep_base_1, prep.Nan_to_mean()]
)

# Rellenado de nan con la moda
prep_3 = prep.Preprocessing(
    [prep_base_1, prep.Nan_to_mode()]
)

# Rellenado de nan con la median
prep_4 = prep.Preprocessing(
    [prep_base_1, prep.Nan_to_median()]
)
```

```python
# Sin la columna fila
prep_base_2 = prep.Preprocessing(
    [prep.Drop_high_cardinals(), prep.Drop_column_nan(), prep.Dummy_variables()]
)

# Rellenado de nan con zeros
prep_5 = prep.Preprocessing(
    [prep_base_2, prep.Nan_to_zero()]
)

# Rellenado de nan con la media
prep_6 = prep.Preprocessing(
    [prep_base_2, prep.Nan_to_mean()]
)

# Rellenado de nan con la moda
prep_7 = prep.Preprocessing(
    [prep_base_2, prep.Nan_to_mode()]
)

# Rellenado de nan con la median
prep_8 = prep.Preprocessing(
    [prep_base_2, prep.Nan_to_median()]
)
```

```python
list_preprocessing = [
    # con la columna fila
    prep_1, prep_2, prep_3, prep_4,
    # sin la columna fila
    prep_5, prep_6, prep_7, prep_8
]
```

## Entrenamos el modelo

### GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
from common import training

models_preprocessing = training(GaussianNB(), X_train, y_train, list_preprocessing)
```

```python
from common import plot_rocs

best_model_index = plot_rocs(proba_predicts, y_val, 4, 2)
```

Se puede observar que el modelo 1 es el que posee mejor auc_roc_score

```python
best_gnb_prep = models_preprocessing[best_model_index]
```

Con el siguiente preprocesado

```python
best_gnb_prep[1].function()
```

A comparacion de los otros modelo este parece tener un score muy bajo

Ahora probamos con el multinomialNB

### MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB

models_preprocessing = training(MultinomialNB(), X_train, y_train, list_preprocessing)
```

```python
proba_predicts = get_proba_predicts(models_preprocessing, X_val)
```

```python
best_model_index = plot_rocs(proba_predicts, y_val, 4, 2)
```

Se puede observar que el mejor modelo es el 6, ademas mejoró significativamente.

```python
best_mnb_prep = models_preprocessing[best_model_index]
```

Con el siguiente preprocesado

```python
best_mnb_prep[1].function()
```

Veamos el categoricalNB

### CategoricalNB

```python
from sklearn.naive_bayes import CategoricalNB

models_preprocessing = training(CategoricalNB(), X_train, y_train, list_preprocessing)
```

```python
proba_predicts = get_proba_predicts(models_preprocessing, X_val)
```

```python
best_model_index = plot_rocs(proba_predicts, y_val, 4, 2)
```

Se puede observar que el mejor modelo de nuevo es el 6 y que mejoró un poco.

```python
best_cnb_prep = models_preprocessing[best_model_index]
```

Con el siguiente preprocesado

```python
best_cnb_prep[1].function()
```

## Escogemos el mejor modelo

```python
models_prep_final = [best_gnb_prep, best_mnb_prep, best_cnb_prep]

proba_predicts = get_proba_predicts(models_prep_final, X_val)
```

```python
best_model_index = plot_rocs(proba_predicts, y_val, 3, 1)
```

Se puede observar que el mejor modelo resulto ser el CategoricalNB.

```python
best_nb, preprocessing = models_prep_final[best_model_index]
```

Con el siguiente preprocesado

```python
preprocessing.function()
```

El modelo mejoro un poco pero sigue siendo un rendimiento bajo en comparacion con los otros notebooks. Esto puede deberse a que el presupuesto que hace naive bayes respecto a que las variables predictoras son independientes entre si, tiene un error alto.
